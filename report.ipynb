```python
from utils import get_file, get_account_info, organize_accounts
from matplotlib import pyplot as plt
import collections

transactions = get_file('data/transactions.json')
```

```python
account_summary = organize_accounts(transactions)
account_numbers = list(account_summary.keys())
account_sums = [account_summary[number]['summed ammounts'] for number in account_numbers]
```

```python
all_locations = []

for account in account_numbers:
    all_locations += account_summary[str(account)]['locations']
```

```python
x = account_numbers
y = account_sums

plt.bar(x, y, color = 'pink')
plt.xticks(rotation = 90)
plt.title('Transaction Amounts For Each Account')
plt.show()
```

```python
location_counts = collections.Counter(all_locations).most_common()
names = [location[0] for location in location_counts]
counts = [location[1] for location in location_counts]
```

```python
x = names[:10]
y = counts[:10]

plt.bar(x, y, color = 'purple')
plt.xticks(rotation = 90)
plt.title('Most Common Locations')
plt.show()
```
